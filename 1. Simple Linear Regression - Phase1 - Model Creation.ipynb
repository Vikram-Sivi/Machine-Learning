```python
# Pandas is a python library used for working and manupulating with datasets.Name "Pandas" references to both "Panel Data", and "Python Data Analysis"
# Importing a library is a one-time action

import pandas as pd
```

```python
# uploading the csv file to variable "dataset" using read_csv function from pandas library
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
# Assigning a seperate column of dataset to a variable
    # Splitting input and output variable

independent=dataset[["YearsExperience"]]
independent
```

```python
dependent=dataset[["Salary"]]
dependent
```

```python
# Splitting train set and test set using *sklearn library*
# Callling train_test_split() function with passing parameters

from sklearn.model_selection import train_test_split
train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# Splitting Train set and Test set

## You can see totally four lists were returned while using sklearn's function. We should assign each
## list to a variable, so that only we can assign them as input for another function
## x_train - Train set input, x_test - Test set input, y_train - Train set o/p, y_test - Test set o/p

### we can use sklearn library only is we seperate data like this. Beacuse the library is set that way
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

### Model Creation
#### ---> Regression is a method for understanding the relationship between independent variables or features and a dependent variable or outcome
--->  A regressor is a statistical term. It refers to any variable in a regression model that is used to predict a response variable.
    We are mentioning here as regressor(variable) for easy understanding, which is followed in most companies

```python
# Using class - LinearRegression from sklearn Lib to create Model
# First two lines here is called as "Procedure"
## Third line - Calling the fit() function inside regressor through Access operator(.), thereby providing I/P & O/P of train set for the Model(AI Brain) to be created

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

The formula to calculate the Slope(w) and Bias(b) of [Y = WX + B] is already coded in class - LinearRegression. Once we gave the training set data the both values were calculated and Model gets created

```python
# Displaying the slope calculated - Assigning to variable 'slope'
#    'coef_' is a function in library which contains formula to calculate slope

slope=regressor.coef_
slope
```

```python
# Displaying the bias value calculated - Assigning to variable 'bias'
#    'intercept_' is a function in library which contains formula to calculate slope

bias=regressor.intercept_
bias
```

### Evaluating metrics of the Model created using test set
##### In Evaluation metrics we will be using the Test data input(X_test) to do prediction (y_pred)

```python
# Note - Reason for Using regressor multiple times is because, all functions(LinearRegression) were imported here.
## predict functions contains formula[Y = WX + b] to calculate the predicted O/P

y_pred=regressor.predict(x_test)
```

```python
# Calculating R2 Value
## Calling 'r2_score' function from library to calulate the respective value
## Passing Test O/P and precticted O/P to calculate R^2 value

from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
# R^2 value ~ 1 ===> Better learned Model
r_score
```

### Saving the Model Created

Pickle is a useful Python tool that allows you to save your ML models, to minimise lengthy re-training and allow you to share, commit, and re-load pre-trained machine learning models. Most data scientists working in ML will use Pickle or Joblib to save their ML model for future use

```python
# Importing pickle and Creating a file to save the model with extension .sav

import pickle
filename="Finalized_model_linear.sav"
```

```python
# Dumping the Model created to the file - Finalized_model_linear.sav
## The python open() function is used to open() internally stored files. It returns the contents of the file as python objects
##    Syntax : open(file_name, mode) 

pickle.dump(regressor,open(filename,'wb'))
```

The Model has been saved. By now you can see the file with name "Finalized_model_linear.sav" is created in your system folder as well.

### Loading the saved Model

```python
# Loading the Model created to a variable - loaded_model, to proceed to Stage-II (Deployment)
# Testing the Model created by providing end user input

loaded_model=pickle.load(open("Finalized_model_linear.sav",'rb'))
result = loaded_model.predict([[15]])

# r : This string is used to read(only) the file. It is passed as default if no parameter is supplied and returns an error if no such file exists.
```
